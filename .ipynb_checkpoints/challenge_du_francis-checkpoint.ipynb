## Loading dataset and General Overview

In the following part we will do some basic exploration of the data to get a feel for it. We start by displaying the first 5 entries and printing some info.

```python
# Load in data 
df = pd.read_json('planning.json')
df.head()
```

Taking a first look, our data consists of 10'000 rows and 20 columns. 
We see that talentName, talentId and talentGrade are empty when isUnassigned is set to True.
For assigned projects the bookingGrade is empty.
We also see that talentName not only includes the full name but could also have titles that indicate gender or educational level. Furthermore, requiredSkills and optionalSkills are nested. We replace empty values with NaN and visualize them in a heat map.

```python
def plot_nan(df):
    cols = df.columns
    colours = ['#8F00FF', '#ffff00'] # yellow means NaN values, violet means non NaN values.
    ax = plt.axes()
    sns.heatmap(df[cols].isna(), cmap=sns.color_palette(colours), ax=ax)

    ax.set_title('NaN values in dataset (1=True, 0=False)')
    plt.show()

#Replace empty values with nan and plot
df = df.replace(r'^\s*$', np.nan, regex=True)
plot_nan(df)
```

```python
# We check how many nan values there are
df.isna().sum()
```

# In-depth statistics#

We give some insights to the industries and the skills. For now, we look at the industries and how many jobs there are and which skills are most common.

## Industries ##
We are interested in seeing how many jobs each industry has.

```python
df_ind = df["industry"].value_counts()

bar = df_ind.plot(kind='bar', figsize=(11,7))
bar.set_title('Number of projects per Industry')
df_ind.head()
```

We can see that not all industries are represented evenly.

```python
df_ind.describe()
```

```python
df_industry = df[['isUnassigned', 'industry']]
industries = []
for industry in df['industry'].value_counts().index:
    for row in df_industry.iterrows():
        if(row[1]['industry']==industry and row[1]['isUnassigned']==True): 
                industries.append(row[1]['industry'])
    counter_industries = Counter(industries)
    keys = list(counter_industries.keys())
    values = list(counter_industries.values())

f = plt.figure()
f.set_figwidth(15)
plt.xticks(rotation='vertical')
plt.ylabel("number of unassigned projects")
plt.bar(keys, values)
```

We see that Music, Textile, Film, Publishing and Floral are the industries with the most jobs with 400 each. We also see that Textile, Floral and Leisure are the industries with the most unassigned projects.

## Skills ##

We are interested in seeing which skills are overall used the most for each talent grade and how they are distributed in the top industries. 

```python
#function for generating new dataframe for skills
def gen_df(s):
    skills_df = df[['id','talentGrade','industry', s]].explode(s)
    skills_df.dropna(axis=0, inplace=True)
    skills_df = pd.concat([skills_df.drop([s], axis=1), skills_df[s].apply(pd.Series)], axis=1)
    return skills_df
    
#We generate dataframes for required skills and optional skills
df_req_skills = gen_df('requiredSkills')
display(df_req_skills)

df_opt_skills = gen_df('optionalSkills')
display(df_opt_skills)
```

```python
#Plot the skills for assigned jobs in a bar plot
def plot_skills(df, title):
    skills = df.groupby('name')['talentGrade'].value_counts().unstack()
    bar = skills.plot.bar(stacked=True, figsize=(11,7))
    bar.set_title(title)
    bar.legend(bbox_to_anchor=(1, 0.5), loc='center left')
    bar.set_ylabel('Number of Jobs')

plot_skills(df_req_skills, 'Required skills for assigned jobs')
plot_skills(df_opt_skills, 'Optional skills for assigned jobs')
```

Now we can look at the top and bottom industries and see which required skills in assigned jobs they use the most

```python
def chart_industry_skills(df, industries):    
    skills= []
    for industry in industries:
        for row in df.iterrows():
            if(row[1]['industry']==industry): 
                    skills.append(row[1]['name'])
        counter_skills = Counter(skills)
        keys = list(counter_skills.keys())
        values = list(counter_skills.values())
        plt.pie(values, labels = keys, autopct='%1.f%%')
        plt.title("Skill Chart for " + industry + " Industry")
        plt.show() 

#We look at the top 5 industries
chart_industry_skills(df_req_skills, df['industry'].value_counts().head().index)
```

```python
#We look at the bottom 5 industries
chart_industry_skills(df_req_skills, df['industry'].value_counts().tail().index)
```

We see that in the top industries the required skills are pretty balanced, whereas in the bottom industries we see some differences.
